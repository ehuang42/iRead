```python
url_list = ['https://www.matataiwan.com/category/mata-%e9%81%b8%e6%9b%b8/', 'https://www.matataiwan.com/category/mata-%e9%81%b8%e6%9b%b8/page/2/']
for i in range(96):
    if i ==0:
        url_list.append('https://www.matataiwan.com/tag/culture/')
    else:
        url_list.append('https://www.matataiwan.com/tag/culture/page/' + str(i+1) + '/')
links_list = []
for url in url_list:
    print('log', url)
    read_html = getHTML(url)
    links_list.append(grab(read_html, 'div', 'grid-header-box'))
```

```python
pact = []
count = 1
for i in links_list:
#     print('new_page_start')
    for index in range(len(i)):
        print(count)
        count += 1
        obj = {}
        title = grab(i[index], 'h2', 'entry-title grid-title')
        link = title[0].a.get('href')
        author = grab(i[index], 'a', 'url fn n')
        date = grab(i[index], 'time', 'entry-date published')
        print('title: ', title[0].text)
        print('author: ', author[0].text)
        print('link: ', link)
        if (date):
            print('date:', date[0].text)
        else:
            print('date not found')
        print('\n')
        obj['title'] = title[0].text
        obj['link'] = link
        obj['author'] = author[0].text
        if (date):
            obj['date'] = date[0].text
        else:
            obj['date'] = 'none'
        pact.append(obj)
    
# print(pact)
```
